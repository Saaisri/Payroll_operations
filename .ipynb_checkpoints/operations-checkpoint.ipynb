```python
import pandas as pd
import numpy as np
```

```python
new_data = pd.read_csv ('new_data.csv')
old_data = pd.read_csv ('old_data.csv')
exp = pd.read_csv ('experience.csv')
```

```python
# Displaying the tables
```

```python
new_data
```

```python
del new_data['City']
del new_data['Hire Date']
```

```python
old_data = old_data.rename(columns={'Hire date':'Hire_date'})
```

```python
old_data
```

```python
exp
```

```python
# Filling the first and last name using split functions
```

```python
new_data['Name'] = new_data['Email'].apply(lambda x: x.split('@')[0])
```

```python
new_data
```

```python
new_data[['First Name','Last Name']] = new_data.Name.str.split(".",expand=True)
```

```python
df = pd.merge(new_data, old_data, on = "Email", how = "outer")
```

```python
df
```

```python
df[['Month','Date','Hiring Year']] = df.Hire_date.str.split("/",expand=True)
```

```python
del df['Date']
del df['Month']
del df['Name']
```

```python
# Calculate the Year of the employee joining
```

```python
df["City"].fillna("London", inplace = True)
df["Hiring Year"].fillna("2020", inplace = True)
df["Hire_date"].fillna("1/5/2020", inplace = True)
```

```python
df["Payroll"] = df["Payroll"].replace("[£,]", "", regex=True)
```

```python
df["Payroll"] = df["Payroll"].astype(int)
```

```python
df['Payroll'] = df['Payroll'].fillna(0)
df['Payroll Raise %'] = df['Payroll Raise %'].fillna(0)
```

```python
df
```

```python
# Calculating the payroll raise based on the year of joining 
```

```python
exp = {
    '2017': 0.01, 
    '2018': 0.02, 
    '2019': 0.03
}

df['Payroll Raise %'] = df['Hiring Year'].map(exp)
```

```python
df["Raise Payroll"] = df["Payroll"] * df["Payroll Raise %"]
```
